# 1.sales_overview

```python
#choose the feature of from dataset 1 for salesoverview 
sale1=ds1[['merchantid', 'totalunitssold', 'meanproductprices']]
#choose the feature of from dataset 2 for salesoverview 
sale2=ds2[['rating', 'units_sold', 'price', 'retail_price']]
#combine the features to get one dataset for sales overview
sales_overview_df=pd.concat([sale1, sale2], axis=1)
sales_overview_df.head()
```

#  2.seller_performance

```python
#choose the feature of from dataset 1 for salesoverview 
saler1=ds1[['merchantid', 'listedproducts', 'totalunitssold', 'meanunitssoldperproduct', 'merchantratingscount', 'meanproductprices', 'totalurgencycount']]
#choose the feature of from dataset 2 for salesoverview 
saler2=ds2[['merchant_rating_count']]
#combine the features to get one dataset for sales overview
seller_performance_df=pd.concat([saler1, saler2], axis=1)
seller_performance_df.head()
```

### The obtained  dataset provides information about the performance of different sellers or merchants

# 3.product_ratings

```python
#choose the feature of from dataset 2 for salesoverview 
rating1=ds2[['title', 'rating', 'rating_count', 'rating_five_count', 'rating_four_count', 'rating_three_count',
       'rating_two_count', 'rating_one_count', 'merchant_rating']]
#choose the feature of from dataset 3 for salesoverview 
rating2=ds3[['tag','count']]
#combine the features to get one dataset for sales overview
product_ratings_df=pd.concat([rating1, rating2], axis=1)
product_ratings_df.head()
```
