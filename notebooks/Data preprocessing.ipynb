# Data preprocessing and data cleaning

```python
#check missing values
p_perfomance_df.isnull().sum()
```

```python
# its seem that all missing values in rating has rating_count 0
# Filter rows where 'rating_count' is equal to zero
filtered_rating = p_perfomance_df[p_perfomance_df['rating_count'] == 0]
filtered_rating[['rating_count','rating_five_count','rating_four_count','rating_three_count','rating_two_count','rating_one_count']]
```

```python
# since the missing value are in rating which is zero we will fill the missing value zero
```

```python
# Remove extra spaces from column names
p_perfomance_df.columns = p_perfomance_df.columns.str.strip()
columns_to_fill = ['rating_five_count', 'rating_four_count', 'rating_three_count', 'rating_two_count', 'rating_one_count']
p_perfomance_df.loc[:, columns_to_fill] = p_perfomance_df.loc[:, columns_to_fill].fillna(0)
p_perfomance_df.isnull().sum()
```

```python
mode_product_color = p_perfomance_df['product_color'].mode()[0]  
p_perfomance_df['product_color'].fillna(mode_product_color, inplace=True)
p_perfomance_df.isnull().sum()
```

```python
tags_column = 'tags'  

# Combine all tags into a single string
all_tags = ','.join(p_perfomance_df[tags_column])

# Split the tags into individual words
individual_tags = all_tags.split(',')

# Create a DataFrame with the tags
df_tags = pd.DataFrame({tags_column: individual_tags})

# Define category rules based on common themes
category_rules = {
    'Summer': ['Summer'],
    'Fashion': ['Fashion', 'Dress', 'Clothing'],
    'Casual': ['Casual', 'Shirt', 'Shorts'],
    'Print': ['Print', 'Flamingo'],
    'Sleepwear': ['Sleepwear', 'Pajamas', 'Nightwear']
}

def assign_category(tag):
    for category, keywords in category_rules.items():
        if any(keyword.lower() in tag.lower() for keyword in keywords):
            return category
    return 'Other'

p_perfomance_df['category'] = p_perfomance_df[tags_column].apply(assign_category)

p_perfomance_df.head()
```
